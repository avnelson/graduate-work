```python
#Q1. min/max/avg price, year, odometer for each car model.
models = vehicles.model.unique()
ModelList = {}
count = 0
for item in vehicles:
    df1 = vehicles[vehicles['model'] == models[count]]
    
    Manuf = df1['manufacturer'].unique()
    
    Qty = len(df1)
    
    MinPrice = min(df1['price'].dropna())
    MaxPrice = max(df1['price'].dropna())
    AvgPrice = df1['price'].dropna().mean()
    
    MinYear = min(df1['year'].dropna())
    MaxYear = max(df1['year'].dropna())
    AvgYear = int(df1['year'].dropna().mean())
    
    MinOdo = min(df1['odometer'].dropna(), default=0) 
    MaxOdo = max(df1['odometer'].dropna(), default=0)
    AvgOdo = df1['odometer'].dropna().mean()
    
    #add calculations as new dict entry
    ModelList[models[count]] = [Manuf, Qty, MinYear, AvgYear, MaxYear, MinPrice, AvgPrice, MaxPrice, MinOdo, AvgOdo, MaxOdo]
    count = count + 1
    
ModelStats = pd.DataFrame.from_dict(ModelList, orient='index')
ModelStats.columns = ['Make', 'Quantity', 'Oldest Year', 'Avg Year', 'Newest Year', 'Cheapest', 'Mid Range', 'Most Expensive', 'Least Miles', 'Avg Mileage', 'Most Miles']

ModelStats.to_csv(r'/Users/avanelson/Desktop/ModelStats.csv', header=True)
ModelStats
```

```python
#FUNCTION TO READ MODEL STATISTICS

def ModelStats(df, model):
    df1 = df[df['model'] == model] #isolate car model you want
    
    #determine the mode of chosen variables
    QtyModel = len(df1)
    
    #get cylinder count mode
    cyl_dict = {}
    count = 0
    for cyl in df1.cylinders:
        if cyl in cyl_dict.keys():
            continue
        elif cyl == 'nan':
            continue
        else:
            cyl_dict[cyl] = len(df1[df1['cylinders'] == cyl])
  
    
    cyl_sorted = [(k, v) for k, v in sorted(cyl_dict.items(), key=lambda x: x[1], reverse=True)]
    Cylinders = cyl_sorted[0][0]
    
    #get color mode
    color_dict = {}
    count = 0
    for color in df1.paint_color:
        if color in color_dict.keys():
            continue
        elif color == 'nan':
            continue
        else:
            color_dict[color] = len(df1[df1['paint_color'] == color])
  
    
    colors_sorted = [(k, v) for k, v in sorted(color_dict.items(), key=lambda x: x[1], reverse=True)]
 
    Color = colors_sorted[0][0]
    
    
    #determine quantity of most frequently occuring
    QtyCylinders = cyl_sorted[0][1]
    QtyColor = colors_sorted[0][1]

    #restructure mode strings to pass into next argument
    Cylinders = Cylinders.replace('0    ','')
    Color = Color.replace('0    ','')

    #isolate DFs further by variable
    CylPrice = df1[df1['cylinders'] == Cylinders]
    ColorPrice = df1[df1['paint_color'] == Color]

    #find average price by variable
    ModelAvgPrice = df1.price.dropna().mean()
    CylAvgPrice = CylPrice.price.dropna().mean()
    ColorAvgPrice = ColorPrice.price.dropna().mean()

    #add calculations as new dict entry
    ModelSumDict = {'Model': [model, QtyModel, ModelAvgPrice],
                'Cylinders': [Cylinders, QtyCylinders, CylAvgPrice],
                'Color': [Color, QtyColor, ColorAvgPrice]}
    ModelStatsDF = pd.DataFrame.from_dict(ModelSumDict, orient='index')
    ModelStatsDF.columns = ['Most Frequent', '# Listings', 'Avg Price']
    print(ModelStatsDF)
    
    #plot year, price, odometer on boxplot
    df2 = df1[['year', 'price', 'odometer']]
    df_sorted = pd.DataFrame(np.sort(df2, axis=0), index=df2.index, columns=df2.columns)
    df_mean = df_sorted.mean(axis=1)
    df_mean.index = np.arange(1, len(df_mean) + 1)
    df_qn =df2.rank(method="min").stack().astype(int).map(df_mean).unstack()
    
    sns.boxplot(data=df_qn)
    
    # set y-axis label
    plt.ylabel("Measurement", size=18)
    plt.title("Boxplot for Selected Vehicle")
    plt.savefig('Boxplot_after_Quantile_Normalization_Seaborn.png',dpi=150)
    

   
```
