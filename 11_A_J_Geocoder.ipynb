```
df = pd.read_csv('a_outliers_locations.csv')
df.head()
```

```
df2 = pd.read_csv('j_outliers_locations.csv')
df2.head()
```

```
#From tutoral on https://towardsdatascience.com/
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
locator = Nominatim(user_agent='myGeocoder')
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
df['location2'] = df['location'].apply(geocode)
df['point'] = df['location2'].apply(lambda loc: tuple(loc.point) if loc else None)
df[['latitude', 'longitude', 'altitude']] = pd.DataFrame(df['point'].tolist(), index=df.index)
df.to_csv('a_outlier_data.csv')
!cp a_outlier_data.csv "drive/My Drive/"
```
